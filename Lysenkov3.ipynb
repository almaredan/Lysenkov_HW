# Сетка в ($\mathrm{\beta}$, R)

```python
plt.vlines(R[0], 0, 2*np.pi, "b")
plt.vlines(R[-1], 0, 2*np.pi, "b")
plt.hlines(0, R[0], R[-1], "b")
plt.hlines(2*np.pi, R[0], R[-1], "b")

for angle in beta:
    plt.hlines(angle, R[0], R[-1], "r", lw=0.5)

for radius in R:
    plt.vlines(radius, 0, 2*np.pi, "r", lw=0.5)
    
plt.xlabel("R")
plt.ylabel(r"$\beta$")
```
