Замена импорта всех инженерных библиотек одним вызовом

Работает только в Jupyter Notebook

```python
%pylab inline
```

```python
def f(x):
    return 2.5*e**(-2*x**2) - 0.5
    
def num_cent(f, x0, h):
    return (f(x0+h/2) - f(x0-h/2))/h

def num_right(f, x0, h):
    return (f(x0+h) - f(x0))/h

def df(x):
    return -10*np.e**(-2*x**2)*x
    
x0 = -10
x1 = 10

h1 = 1
h2 = 0.1

x = np.arange(x0, x1+h2, h2)
_x = np.arange(x0, x1+h1, h1)
```

```python
plt.figure(figsize=(20, 12))
#plt.plot(x, f(x), '-g', label=u'Func')
plt.plot(np.arange(x0,x1,0.01), df(np.arange(x0,x1,0.01)), '-b', label=u'Derivative')
plt.plot(_x, num_cent(f, _x, h1), '--xr', label=u'Numerical Central Diff h=1')
plt.plot(_x, num_right(f, _x, h1), '--xg', label=u'Numerical Right Diff h=1')
plt.legend(loc=4)
plt.xlim(-3,3)
plt.show()
```

```python
plt.figure(figsize=(20, 12), )
#plt.plot(x, f(x), '-g', label=u'Func')
plt.plot(np.arange(x0,x1,0.01), df(np.arange(x0,x1,0.01)), '-b', label=u'Derivative')
plt.plot(x, num_cent(f, x, h2), '--xr', label=u'Numerical Central Diff h=0.1')
plt.plot(x, num_right(f, x, h2), '--xg', label=u'Numerical Right Diff h=0.1')
plt.legend(loc=4)
plt.title('Derivative')
plt.xlim(-3, 3)
plt.show()
```
